```python
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd
documents = [D1, D2, D3, D4,D5,D6,D7,D8,D9,D10,D11,D12,D13,D14,D15,D16,D17,D18,D19,D20,D21,D22,D23,D24,D25,D26,D27,D28,D29,D30]
# Initialize TfidfVectorizer
vectorizer = CountVectorizer(max_df=0.95, min_df=1, stop_words="english")

# Fit and transform the documents
bow = vectorizer.fit_transform(documents)

# Get feature names (words)
feature_names = vectorizer.get_feature_names_out()

# Convert to DataFrame for better readability
df_bow = pd.DataFrame(bow.toarray(), columns=feature_names)

print("Bag of Words (TF-IDF) Representation:")
print(df_bow)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd

# Calculate cosine similarity between documents
cosine_sim_matrix = cosine_similarity(df_bow)

# Convert to DataFrame for better readability
df_cosine_sim = pd.DataFrame(cosine_sim_matrix, index=documents, columns=documents)

print("Cosine Similarity Matrix:")
print(df_cosine_sim)
```

```python
from sklearn.metrics import jaccard_score
import numpy as np
import pandas as pd

# Convert BOW matrix to binary (presence/absence) for Jaccard similarity
binary_bow = (df_bow > 0).astype(int)

# Calculate Jaccard Similarity
num_documents = binary_bow.shape[0]
jaccard_sim_matrix = np.zeros((num_documents, num_documents))

for i in range(num_documents):
    for j in range(num_documents):
        # For Jaccard similarity, we can use the jaccard_score which computes 1 - Jaccard distance
        # However, jaccard_score expects 1D arrays for binary classification tasks.
        # To compute similarity between two sets represented as binary vectors,
        # we can calculate (intersection_size / union_size)

        # Get the binary vectors for documents i and j
        vec_i = binary_bow.iloc[i]
        vec_j = binary_bow.iloc[j]

        # Calculate intersection and union
        intersection = np.sum(np.logical_and(vec_i, vec_j))
        union = np.sum(np.logical_or(vec_i, vec_j))

        if union == 0:
            jaccard_sim_matrix[i, j] = 0.0 # If both sets are empty, similarity is 0
        else:
            jaccard_sim_matrix[i, j] = intersection / union

# Convert to DataFrame for better readability
df_jaccard_sim = pd.DataFrame(jaccard_sim_matrix, index=documents, columns=documents)

print("Jaccard Similarity Matrix:")
print(df_jaccard_sim)
```

```python
import nltk
from nltk.corpus import wordnet
from nltk.tokenize import word_tokenize
import numpy as np
import pandas as pd

# Download required NLTK resources
try:
    nltk.data.find('corpora/wordnet')
except LookupError:
    nltk.download('wordnet')

try:
    nltk.data.find('tokenizers/punkt')
except LookupError:
    nltk.download('punkt')

# Define the 30 documents
D1="The cat is sleeping on the mat"
D2="She enjoys reading books in the evening"
D3="The sun rises in the east every morning"
D4="I like to drink coffee before breakfast"
D5="The children are playing in the park"
D6="He went to the store to buy groceries"
D7="They are watching a movie together"
D8="My favorite color is blue"
D9="We will meet at the library at noon"
D10="The dog barked loudly at the stranger"
D11="The mat is where the cat is taking a nap"
D12="In the evening, she likes to read books"
D13="Each morning, the sun comes up in the east"
D14="Before breakfast, I prefer drinking coffee"
D15="In the park, the kids are having fun playing"
D16="He visited the store to get some groceries"
D17="Together, they are enjoying a film"
D18="Blue is the color I like the most"
D19="At twelve o’clock, we will gather in the library"
D20="The stranger was greeted with loud barks from the dog"
D21="I bought a new laptop yesterday"
D22="The train arrives at the station at 5 PM"
D23="My phone battery died while I was traveling"
D24="Elephants are the largest land animals"
D25="I need to finish my homework by tonight"
D26="Space exploration requires advanced technology"
D27="I learned how to play the guitar last summer"
D28="The weather forecast predicts heavy rain tomorrow"
D29="She baked a chocolate cake for her friend’s birthday"
D30="Mount Everest is the tallest mountain in the world"

# Prepare documents list
documents = [D1, D2, D3, D4, D5, D6, D7, D8, D9, D10,
             D11, D12, D13, D14, D15, D16, D17, D18, D19, D20,
             D21, D22, D23, D24, D25, D26, D27, D28, D29, D30]

# Function to compute WordNet similarity
def document_wordnet_similarity(doc1, doc2):
    tokens1 = word_tokenize(doc1.lower())
    tokens2 = word_tokenize(doc2.lower())

    synsets1 = [s for token in tokens1 for s in wordnet.synsets(token)]
    synsets2 = [s for token in tokens2 for s in wordnet.synsets(token)]

    if not synsets1 or not synsets2:
        return 0.0

    # Calculate pairwise similarity and take the maximum for each synset
    max_similarities = []
    for s1 in synsets1:
        max_sim_for_s1 = 0.0
        for s2 in synsets2:
            sim = s1.path_similarity(s2)
            if sim is not None and sim > max_sim_for_s1:
                max_sim_for_s1 = sim
        max_similarities.append(max_sim_for_s1)

    return np.mean(max_similarities) if max_similarities else 0.0

# Calculate WordNet similarity matrix
num_documents = len(documents)
wordnet_sim_matrix = np.zeros((num_documents, num_documents))

for i in range(num_documents):
    for j in range(num_documents):
        if i == j:
            wordnet_sim_matrix[i, j] = 1.0
        else:
            wordnet_sim_matrix[i, j] = document_wordnet_similarity(documents[i], documents[j])

# Convert to DataFrame for readability
df_wordnet_sim = pd.DataFrame(wordnet_sim_matrix, index=[f"D{i+1}" for i in range(num_documents)],
                              columns=[f"D{i+1}" for i in range(num_documents)])

print("WordNet Similarity Matrix (Path Similarity):")
print(df_wordnet_sim)
```
